```python
cardio.sort_values(by=['val'], ascending=False) # max колл.смертей, от заболиваний сердца
# .sort_values(): Метод для сортировки данных
# by=['val']: Параметр указывает столбец для сортировки
# ascending=False: Параметр определяет направление сортировки
```

```python
df.drop('measure', axis=1, inplace=True)
# df.drop(...): Метод для удаления строк или столбцов
# 'measure': Имя столбца для удаления
# axis=1: Параметр указывает, что мы удаляем столбец .. 0 это значит строку 
# inplace=True: Параметр указывает, что изменения должны применяться к исходному DataFrame
```

```python
df.drop(['metric','age','upper','lower'], axis=1, inplace=True) # удалили все не нужные столбцы
```

```python
df['val_round'] = df['val'].round(decimals=1) 
# df['val_round'] тут создаем новый столбец 
# = df['val'].round(decimals=1)  и тут же в него запихиваем значения другого столбца и округлаем показатель 
```

```python
df.rename(columns= {'val':'value'}, inplace=True) # переменовали inplace=True в исходом DF
```
